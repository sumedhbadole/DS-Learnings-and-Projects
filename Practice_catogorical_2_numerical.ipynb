```python
#OrdinalEncoding
```

```python
df = pd.read_csv('customer.csv')
df.head()
```

```python
ordered_columns = df.iloc[: , 2:4]
ordered_columns.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
sc=OrdinalEncoder()
```

```python
Result=sc.fit_transform(ordered_columns)
```

```python
Result=pd.DataFrame(Result,columns=['reviews','education_1'])
```

```python
df=pd.concat((df,Result),axis=1)
df
```

```python
df
```

```python
df=df.drop(['review','education'],axis=1)
```

```python
df
```
