```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
```

```python
# Create the dataset
data = {
    'House Area': [1000, 1500, 2000, 2500, 3000],
    'House Price': [250, 400, 450, 550, 600]
}

# Convert the dataset to a pandas DataFrame
df = pd.DataFrame(data)
df
```

```python
# Separate the independent (x) and dependent (y) variables
x = df[['House Area']]
y = df['House Price']
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x, y)

slope = model.coef_[0]
intercept = model.intercept_
```

```python
# Make predictions using the model
predicted_prices = model.predict(x)
```

```python
# Visualize the data and regression line
plt.scatter(x, y, label='Actual Prices',color='g')
plt.plot(x, predicted_prices, color='red', label='Regression Line')
plt.xlabel('House Area (sq. ft)')
plt.ylabel('House Price (thousands of dollars)')
plt.title('Simple Linear Regression: House Area vs. House Price')
plt.legend(loc=2)
plt.show()
```

```python
new_house_area = 1800
predicted_price_new = model.predict([[new_house_area]])
print(f'Predicted House Price for {new_house_area} sq. ft: {predicted_price_new[0]:.2f} thousands of dollars')
```
