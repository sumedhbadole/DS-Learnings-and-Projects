# what is the difference between R2_score and Adjusted r2_score

```python
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sn
```

```python
df=pd.read_csv('Case2.csv')
```

```python
df.head()
```

```python
df
```

```python
x=df[['YearsExperience']]
y=df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x,y)

Pred=reg.predict(x)

Pred
```

```python
df['Pred']=Pred
df
```

```python
from sklearn.metrics import r2_score
```

```python
print('The R2_score(Accuracy) of the model will be',r2_score(y,Pred))
```

```python
1 - (1 - r2_score(y,Pred)) * (len(df)-1) / (len(df) - df.shape[1] - 1)
```
